This example shows how to plot a simple dataset with vline using resonator data.

## Step 1: Import packages 
We import both the Plotter and os. load_labber_file is required to exstrate the data from Labber files.

```python
from dataanalyzer import Plotter, load_labber_file
import os
```

# Step 2: Load the data
We load the data using os and load_labber_file. If inspect = True in load_labber_file, the parameter names will be automatically printed.

```python
base_path = os.path.abspath(os.path.join(os.getcwd() ,"../../"))
path = os.path.join(base_path, "sample_data/Data_Resonator.hdf5")

freq, s21 = load_labber_file(path, insepct=True)
```

# Step 3: Initalize Plotter
And add the data.

```python
# Initialize plotter
plot = Plotter()

# Add data
plot.plot(freq, s21.abs.db, label="Data")

# Plot minimum value of s21 as a vertical line
min_s21 = s21.abs.argmin()
min_freq = freq[min_s21].v[0]

# Add vertical line
plot.axvline(min_freq, label=f"Min value {min_freq:.6} GHz")

# Show plot
plot.show()
```
