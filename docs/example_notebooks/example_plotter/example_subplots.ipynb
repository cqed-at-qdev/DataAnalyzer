This example shows how to plot multiple subplots using the Plotter package. It takes T1 data and plots heatmap along with an average of the traces.

## Step 1: Import packages 
We import both the Plotter and os. load_labber_file is required to exstrate the data from Labber files.

```python
from dataanalyzer import Plotter, load_labber_file
import os
```

# Step 2: Load the data
We load the data using os and load_labber_file. If inspect = True in load_labber_file, the parameter names will be automatically printed.

```python
z_drive = r"Z:\Malthe\DataAnalyzer\sample_data"
path = os.path.join(z_drive, "Data_T1_Repetition.hdf5")

[duration, repetition], [pulse, _] = load_labber_file(path, insepct=True)
```

# Step 3: Initalize Plotter
And add the data.

```python
# Initialize plotter
plot = Plotter(subplots=(2, 1))

# Add data (first subplot)
plot.plot(
    duration,
    pulse.abs.db.mean(axis=0),
    label=f"Mean data ({len(repetition.v)} averages)",
)
plot.heatmap(duration, repetition, pulse.abs.db, ax=(1, 0))

# Show plot
plot.show()
```
